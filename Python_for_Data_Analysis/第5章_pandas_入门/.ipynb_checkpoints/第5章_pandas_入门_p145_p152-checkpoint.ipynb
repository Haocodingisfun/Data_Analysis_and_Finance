#### 相关系数与协方差

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
import tushare as ts

all_data = {}
for ticker in ['000001', '000002', '000004', '000005']:
    all_data[ticker] = ts.get_k_data(ticker, start='2017-01-01', end='2017-06-30')[['date', 'close', 'volume']]
```

```python
all_data
```

```python
price = DataFrame({tic: data['close']
          for tic, data in all_data.items()})
```

```python
price
```

```python
price.index = all_data['000001']['date']
```

```python
price
```

```python
price = DataFrame({tic: data["close"].values[:50] for tic, data in all_data.items()}, 
                  index=np.array(all_data['000001']['date'][:50]))
```

```python
volume = DataFrame({tic: data["volume"].values[:50] for tic, data in all_data.items()}, 
                  index=np.array(all_data['000001']['date'][:50]))
```

```python
price
```

```python
volume
```

```python
returns = price.pct_change()
```

```python
returns.tail()
```

```python
# 相关系数
returns['000001'].corr(returns['000002'])
```

```python
# 协方差
returns['000001'].cov(returns['000004'])
```

```python
returns.corr()
```

```python
returns.cov()
```

```python
# 传入一个 Series 将会返回一个相关系数值 Series
returns.corrwith(returns['000001'])
```

```python
# 传入一个 DataFrame 将会计算按列名配对的相关系数
# axis=1，则按行进行计算
returns.corrwith(volume)
```

```python
# thresh=3，会在一行中至少有 3 个非 NA 值时将其保留
df.dropna(thresh=3)
```
