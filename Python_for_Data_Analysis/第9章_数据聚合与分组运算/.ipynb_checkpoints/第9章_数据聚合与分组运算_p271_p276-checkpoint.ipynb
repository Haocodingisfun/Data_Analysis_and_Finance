# 数据聚合

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = DataFrame({'data1': np.random.randn(5) * 2,
                'data2': np.random.randn(5) * 2,
                'key1': ['a', 'a', 'b', 'b', 'a'],
                'key2': ['one', 'two', 'one', 'two', 'one']})
```

```python
df
```

```python
grouped = df.groupby('key1')
```

```python
grouped['data1'].quantile(0.9)
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
```

```python
grouped.agg(peak_to_peak)
```

```python
grouped.describe()
```

```python
df.median()
```

```python
df.prod()
```

```python
tips = pd.read_csv('/Users/wonderful/Desktop/tips.csv')
```

```python
tips
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips['tip_pct']
```

## 面向列的多函数应用

```python
grouped = tips.groupby(['sex', 'smoker'])
```

```python
grouped
```

```python
grouped_pct = grouped['tip_pct']
```

```python
grouped_pct.agg('mean')
```

```python
grouped_pct.agg(['mean', 'std', peak_to_peak])
```

```python
grouped_pct.agg([('foo', 'mean'), ('bat', np.std)])
```

```python
functions = ['count', 'mean', 'max']
```

```python
result = grouped['tip_pct', 'total_bill'].agg(functions)
```

```python
result
```

```python
result['tip_pct']
```

```python
ftuples = [('Durchschnitt', 'mean'), ('Abweuchung', np.var)]
```

```python
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
grouped.agg({'tip': np.max, 'size': 'std'})
```

```python
grouped.agg({'tip_pct': ['min', 'max', 'mean', 'std'],
             'size': 'sum'})
```

# 以“无索引”的形式返回聚合数据

```python
tips.groupby(['sex', 'smoker'], as_index=False).mean()
```
