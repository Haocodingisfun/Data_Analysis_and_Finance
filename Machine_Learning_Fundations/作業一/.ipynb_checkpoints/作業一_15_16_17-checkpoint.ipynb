```python
import pandas as pd
import numpy as np
import random
```

```python
data = np.fromfile('hw1_15_train.dat', sep=' ', dtype=float)
```

```python
data
```

```python
df = pd.DataFrame(index=range(int(len(data)/5)), columns=['x0', 'x1', 'x2', 'x3', 'x4', 'y'])
```

```python
df['x0'] = 1.0
df['x1'] = data[::5]
df['x2'] = data[1::5]
df['x3'] = data[2::5]
df['x4'] = data[3::5]
df['y'] = data[4::5]
```

```python
df
```

```python
W = [.0, .0, .0, .0, .0]
time = 0
for i in df.index:
    score = sum(df.iloc[i, :5] * W)
    # 不能用 np.sign
    if score <= 0:
        sign = -1
    else:
        sign = 1
    if sign != df.iloc[i, 5]:
        time += 1
        W = W + df.iloc[i, :5] * df.iloc[i, 5]
print('time:', time)
print(W)
```

```python
time_list = []
W_list = []
for j in range(2000):
    print('第%s次乱序' % j)
    random.seed = j
    order = random.sample(range(400),400)
    W = [.0, .0, .0, .0, .0]
    time = 0
    for i in order:
        score = sum(df.iloc[i, :5] * W)
        # 不能用 np.sign
        if score <= 0:
            sign = -1
        else:
            sign = 1
        if sign != df.iloc[i, 5]:
            time += 1
            W = W + df.iloc[i, :5] * df.iloc[i, 5]
    print('time: %s' % time)
    time_list.append(time)
    W_list.append(W)
print('time_avg:', np.array(time_list).mean())
```

# 17

```python
time_list = []
W_list = []
for j in range(2000):
    print('第%s次乱序' % j)
    random.seed = j
    order = random.sample(range(400),400)
    W = [.0, .0, .0, .0, .0]
    time = 0
    for i in order:
        score = sum(df.iloc[i, :5] * W)
        # 不能用 np.sign
        if score <= 0:
            sign = -1
        else:
            sign = 1
        if sign != df.iloc[i, 5]:
            time += 1
            W = W + df.iloc[i, :5] * df.iloc[i, 5] * 0.5
    print('time: %s' % time)
    time_list.append(time)
    W_list.append(W)
print('time_avg:', np.array(time_list).mean())
```
